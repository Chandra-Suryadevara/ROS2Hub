## ROS ##
(Robot Operating System) is an open-source framework for developing robot software, providing tools and libraries for building and running robotic applications. It facilitates communication between different parts of a robot system, supports modularity, and is widely used in various applications, including research, industrial automation, and autonomous vehicles

## Features of ROS
* Data Distribution Service
      - A communication pipeline interface with all of our code
  

## Publisher Subscriber Method:
The first is through a publisher subscriber method.

Let's say we have two nodes for our robot. The camera node wants to make available images from our robot's camera, so it publishes these images to anyone who wants to listen in or subscribe to this particular data.



## Services Method:

Let's say we have a survey node, which is a service client that sends a request to turn the robot to turn it's camera 45 degrees.
This request is received by the camera mover node, which is a service server, after completing the request, it will send back a response.

In this example, our response is an image the camera took after turning the camera to the desired angle.



## Actions 
are designed with a three-part architecture consisting of goal, feedback, and result, making them ideal for long-running tasks. This architecture is frequently used in robot navigation. For example, in a warehouse, an operator might send an Autonomous Mobile Robot (AMR) to a specific aisle by setting the destination as the goal in a ROS action. As the AMR travels, it provides feedback to the operator about its progress along the route. If the operator needs to change the destination, they can cancel the initial action and initiate a new one, redirecting the AMR to a different aisle. When the AMR successfully reaches its destination, it sends a final confirmation to the operator, indicating that the task is complete.
